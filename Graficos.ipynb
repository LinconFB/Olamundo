```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)# posiciona o array x no eixo x
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y, marker = 'o', linestyle = '-', color = 'b')
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Grafico de linha')
plt.grid(True)
plt.show()
```

```python
categorias = ['A','B','C','D','E','F']#Horizontal
valores = [10,20,30,70,30,20]#Vertical
plt.bar(categorias, valores)
plt.xlabel('Categorias')
plt.ylabel('Valores')
plt.title('Grafico de Barras')
plt.grid(False)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.scatter(x,y)
plt.xlabel('Valores de X')
plt.ylabel('Valores de Y')
plt.title('Grafico de linha')
plt.grid(True)
plt.show()
```
